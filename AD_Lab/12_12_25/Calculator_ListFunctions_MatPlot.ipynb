```python
import math

def calculator():
    print(f"Available options: ")
    print(f"1. sum")
    print(f"2. Difference")
    print(f"3. Product")
    print(f"4. Quotient")
    print(f"5. Remainder")
    print(f"6. Sine")
    print(f"7. Cosine")
    print(f"8. Tangent")
    print(f"9. exit")

    n = int(input("Input your choice: "))
    if (n >= 1 and n <= 5):
        a = int(input("Input first operand: "))
        b = int(input("Input second operand: "))
        if(n == 1):
            print(a+b)
        elif(n == 2) :
            print(a-b)
        elif(n == 3):
            print(a*b)
        elif(n == 4):
            print(a/b)
        else:
            print(a%b)

    elif(n > 5 and n < 9):
        num = int(input("Input a operand: "))
        if(n == 6):
            print(math.sin(math.radians(num)))
        elif(n == 7):
            print(math.cos(math.radians(num)))
        else:
            print(math.tan(math.radians(num)))

    elif(n == 9):
        return

    else:
        print(f"Input a valid choice")
        return calculator()
```

```python
calculator()
```
