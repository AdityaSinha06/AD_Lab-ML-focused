```python
#lambda functions: (Anonymous functions)
# lambda arguements : expression
# used when : 
# 1. function is small & used once  
# eg: print((lambda x : x*x)(5))
# 2. lambdas are made for map , filter , reduce ::> coz these fns expect another fn as inputs
```

```python
nums = [1,2,3,4]
```

```python
newList = list(map(lambda x : x*2, nums))
```

```python
newList
```

```python
newList = list(filter(lambda x : x%2==0 , nums))
```

```python
newList
```
