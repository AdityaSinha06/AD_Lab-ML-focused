```python
def checkPalindrome(num):
    temp = num
    reversedNum = 0
    while (temp != 0):
        reversedNum = reversedNum*10 + temp%10
        temp //= 10

    return reversedNum == num

num = int(input())
print(checkPalindrome(num))
```

```python
# ArmsstrongNo : 153 = 1^3 + 5^3 + 3^3 :; ^(#digits)
def checkArmstrong(num):
    temp = num
    digiCnt = 0
    while(temp != 0):
        digiCnt += 1
        temp //= 10

    temp = num
    sumNum = 0
    while(temp != 0):
        sumNum += (temp%10)**digiCnt
        temp //= 10

    return num == sumNum
num = int(input())
print(checkArmstrong(num))
```

```python
import math
def isPrime(num):
    if(num <= 1): return False
    if(num == 2 or num == 3): return True
    if(num%2 == 0 or num%3 == 0): return False
    i = 5
    while (i <= math.sqrt(num)):
        if((num%i == 0) or (num%(i+2) == 0)):
            return False
        i += 6

    return True

def allPrimes(a , b):
    for i in range(a , b+1):
        if(isPrime(i)):
            print(i)
            
allPrimes(1 , 100)
```

```python
def evenDigiSum(num):
    sumDigi = 0
    while(num != 0):
        lastDigi = num%10
        if(lastDigi%2 == 0):
            sumDigi += lastDigi

        num //= 10
        
    return sumDigi
    
def oddDigiSum(num):
    sumDigi = 0
    while (num != 0):
        lastDigi = num%10
        if(lastDigi%2 == 1):
            sumDigi += lastDigi

        num //= 10
        
    return sumDigi
        
num = int(input())
print(oddDigiSum(num) , evenDigiSum(num))
```

```python
# perfect num : sum of all proper divisors = num
import math
def isPerfectNum(num):
    divi = 1
    sumDivs = 0
    while (divi < math.sqrt(num)):
        if(num%divi == 0):
            sumDivs += divi
            if(divi != 1):
                sumDivs += (num // divi)
        divi += 1

    return sumDivs == num
    
num = int(input())
print(isPerfectNum(num))
```

```python
# --------------SECTION-C-----------------
def isVowel(ch):
    if(ch == 'a' or ch == 'e' or ch == 'i' or ch == 'o' or ch == 'u'):
        return True
    if(ch == 'A' or ch == 'E' or ch == 'I' or ch == 'O' or ch == 'U'):
        return True
    return False
    
def vowelsCnt(s):
    cnt = 0
    for ch in s:
        if(isVowel(ch)):
            cnt += 1

    return cnt
    
s = input()
print(vowelsCnt(s))
```

```python
def reverseStr(s):
    arr = list(s)
    n = len(arr)
    i = n-1
    rev = []
    while(i>=0):
        rev.append(s[i])
        i -= 1
    print(''.join(rev))
s = input()
reverseStr(s)
```

```python
def checkPalindrome_str(s):
    arr = list(s)
    n = len(arr)
    i = n-1
    rev = []
    while(i>=0):
        rev.append(s[i])
        i -= 1

    return (''.join(rev) == s)
    
s = input()
print(checkPalindrome_str(s))
```

```python
import re

def normalize_whitespace(text: str) -> str:
    return re.sub(r'\s+', ' ', text).strip()

def clean_text(text: str) -> str:
    text = text.lower()
    text = re.sub(r"[^a-z0-9\s']", " ", text)  # keep apostrophes
    text = normalize_whitespace(text)
    return text


if __name__ == "__main__":
    sample = "This is a Sample, text!!! With EXTRA   spaces."
    print("RAW:", sample)
    print("CLEANED:", clean_text(sample))
```
