```python
def reversedString(str):
    n = len(str)
    i = 0 
    j = n-1
    while i < j:
        temp = str[i]
        str[i] = str[j]
        str[j] = temp
        i += 1
        j -= 1

    return str

print(reversedString("Aditya"))
```

```python
def reversedString(str):
    str_list = list(str)
    n = len(str_list)
    i = 0
    j = n-1

    while i<j:
        str_list[i] , str_list[j] = str_list[j] , str_list[i]
        i += 1
        j -= 1

    return ''.join(str_list)

print(reversedString("Aditya"))
```

```python
def isVowel(ch):
    if(ch == 'A' or ch == 'E' or ch == 'I' or ch == 'O' or ch == 'U'):
        return True
    elif(ch == 'a' or ch == 'e' or ch == 'i' or ch == 'o' or ch == 'u'):
        return True
    else:
        return False
        
def vowelsCnt(str):
    n = len(str)
    cnt = 0
    for i in range(0 , n):
        if(isVowel(str[i])):
            cnt += 1

    return cnt

print(vowelsCnt("Aditya"))
```

```python
def factorial(num):
    ans = 1
    for i in range(2 , num+1):
        ans *= i

    return ans

print(factorial(5))
```

```python
def checkPrime(num):
    if(type(num) != int): return False
    if(num < 2): return False

    i = 2
    while i < num:
        if(num % i == 0): return False
        i += 1

    return True

print(checkPrime(10))
```

```python
import math

def checkPrime_Optimised(num):
    if(type(num) != int): return False

    if(num < 2): return False
    if(num == 2 or num == 3): return True
    if(num%2 == 0 or num%3 == 0): return False
    
    i = 5
    while i <= math.sqrt(num):
        if(num % i == 0 or num % (i+2) == 0): return False
        i += 6

    return True

print(checkPrime_Optimised(11))
```
